```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Re-importing with proper handling of the CSV format
data = pd.read_csv('sample_population.csv')

# Display the columns to understand the structure better
data.columns
```

```python
data.head(266)
```

```python

# List of countries for the bar chart (same as previous example)
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Filter the data for these countries and the year 2022 (or the most recent available year)
year = '2022'  # Adjust if another year is more relevant
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

# Bar Chart for Population in 2022
plt.figure(figsize=(10, 6))
plt.bar(filtered_data['Country Name'], filtered_data[year])
plt.xlabel('Country Name')
plt.ylabel(f'Population in {year}')
plt.title(f'Population of Selected Countries in {year}')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Histogram for Population Distribution in 2022
plt.figure(figsize=(10, 6))
plt.hist(filtered_data[year], bins=10, edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Histogram of Population in {year}')
plt.show()
```
